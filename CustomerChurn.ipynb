```python
import pandas as pd
```

```python
df = pd.read_csv('https://msi.martial.co.ke/datasets/Churn.csv')
df
```

```python
# Remove the CustomerID Column
df.drop('customerID', axis=1, inplace=True)
df
```

```python
# MACHINE LEARNING ALGORITHMS are mathematical based, therefore they cannot learn from text Data
# There is a need to perform Label Encoding to converts text to Numerical representation
df.isnull().sum()
```

```python
# LABEL ENCODING
# 1. gender
df.groupby('gender').size()
```

```python
# Female -> 1
# Male -> 0

df['gender'] = df['gender'].replace({
    'Female': 1,
    'Male': 0
})
df
```

```python
# 2. Dependents
df.groupby('Dependents').size()
df['Dependents'] = df['Dependents'].replace({
    'Yes': 1,
    'No': 0
})

df
```

```python
# 3. PhoneService
df.groupby('PhoneService').size()
df['PhoneService'] = df['PhoneService'].replace({
    'Yes': 1,
    'No': 0
})
df
```

```python
# 3. MultipleLines
df.groupby('MultipleLines').size()
df['MultipleLines'] = df['MultipleLines'].replace({
    'Yes': 1,
    'No': 0,
    'No Phone': 2
})

df
```

```python
columns_to_encode = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']
print(df[columns_to_encode].nunique())
for column in columns_to_encode:
    df[column] = df[column].replace({
        'Yes': 1,
        'No': 0,
        'No Internet': 2
    })

df
```

```python
# 3. InternetService
df.groupby('InternetService').size()
df['InternetService'] = df['InternetService'].replace({
    'DSL': 1,
    'Fiber': 0,
    'No': 2
})

df
```

```python
# 3. Contract
df.groupby('Contract').size()
df['Contract'] = df['Contract'].replace({
    'Month': 1,
    'One year': 0,
    'Two year': 2
})

df
```

```python
# 3. PaymentMethod
df.groupby('PaymentMethod').size()
df['PaymentMethod'] = df['PaymentMethod'].replace({
    'Bank': 1,
    'Credit card': 0,
    'Electronic': 2,
    'Mailed' : 3
})

df
```
